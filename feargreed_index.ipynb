# CNN Fear & Greed Index

A Python library for fetching and analyzing CNN's Fear & Greed Index - a popular indicator of market sentiment.

```python
# Install the package (uncomment if needed)
# !pip install fear-greed-index

from fear_greed_index import CNNFearAndGreedIndex
```

```python
# Fetch current Fear & Greed data
fgi = CNNFearAndGreedIndex()

# Display the complete report
print(fgi.get_complete_report())
```

```python
# Access individual values
print(f"Current Score: {fgi.score:.1f}")
print(f"Current Rating: {fgi.rating}")
print(f"Previous Close: {fgi.previous_close:.1f}")
print(f"1 Week Ago: {fgi.previous_1_week:.1f}")
print(f"1 Month Ago: {fgi.previous_1_month:.1f}")
print(f"1 Year Ago: {fgi.previous_1_year:.1f}")
```

```python
# Access individual indicators
print("Individual Indicators:")
print("-" * 50)
for indicator in fgi.all_indicators:
    print(f"{indicator.name}: {indicator.score:.1f} ({indicator.rating})")
```

```python
# Plot historical Fear & Greed Index
import matplotlib.pyplot as plt

fig = fgi.plot_all_charts()
plt.show()
```

```python
# Access historical data for custom analysis
from datetime import datetime

historical = fgi.get_historical_data()
print(f"Historical data points: {len(historical)}")
print("\nLast 5 data points:")
for point in historical[-5:]:
    date = datetime.fromtimestamp(point['x'] / 1000).strftime('%Y-%m-%d')
    print(f"  {date}: {point['y']:.1f} ({point['rating']})")
```
